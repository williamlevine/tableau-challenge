```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from geopy.distance import geodesic
```

```python
january_df = pd.read_csv("Resources/JC-202401-citibike-tripdata.csv")
january_df = january_df.dropna()
january_df.head()
```

```python
july_df = pd.read_csv("Resources/JC-202407-citibike-tripdata.csv")
july_df = july_df.dropna()
july_df.head()
```

```python
def calculate_distance(row):
    start = (row['start_lat'], row['start_lng'])
    end = (row['end_lat'], row['end_lng'])
    return geodesic(start, end).kilometers
```

```python
january_df['distance_km'] = january_df.apply(calculate_distance, axis=1)
january_df.head()
```

```python
july_df['distance_km'] = july_df.apply(calculate_distance, axis=1)
july_df.head()
```

```python
january_df.to_csv("january_data.csv", index=False)
```

```python
july_df.to_csv("july_data.csv", index=False)
```

```python
full_df = pd.concat([january_df, july_df], ignore_index=True)
full_df.head()
```

```python
full_df.info()
```

```python
full_df['geometry'] = full_df.apply(lambda row: Point(row['start_lng'], row['start_lat']), axis=1)

gdf = gpd.GeoDataFrame(full_df, geometry='geometry')

# Load the GeoJSON or shapefile with zip code boundaries
zip_codes = gpd.read_file("Resources/jerseycityzipcodegeojson.geojson")
# zip_codes.head()

# Ensure both GeoDataFrames use the same coordinate reference system (CRS)
gdf = gdf.set_crs(zip_codes.crs, allow_override=True)

# Perform the spatial join
gdf_with_zip = gpd.sjoin(gdf, zip_codes[['zcta5ce10', 'geometry']], how='left', op='within')
gdf_with_zip = gdf_with_zip.rename(columns={"zcta5ce10": "zip_code"}).drop(columns=["index_right"])

# Convert back to a regular DataFrame
df_with_zip = pd.DataFrame(gdf_with_zip.drop(columns="geometry"))
df_with_zip.head()
```

```python
df_with_zip.info()
```

```python
df_with_zip.to_csv("stacked_data.csv", index=False)
```
